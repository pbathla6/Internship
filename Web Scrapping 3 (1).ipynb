```python
driver.get("https://www.amazon.in")
```

```python
role1=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
role1.send_keys('Guitar')

role2=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
role2.click()
```

```python
page_url=[]
start = 0
end = 3
for page in range(start,end):
    
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        page_url.append(i.get_attribute('href'))
next_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
next_button.click()
```

```python
len(page_url)
```

```python
Brand=[]
Title=[]
Price=[]
Replacement=[]
Stock=[]
Delivery=[]
```

```python

for i in page_url:
    driver.get(i)
    
    try:
        title=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Title.append(title.text)
    except NoSuchElementException:
        Title.append('-')
        
    try:
        brand=driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')
        
        
    try:
        price=driver.find_element(By.XPATH,'//div[@id="corePrice_feature_div"]')
        Price.append(price.text.split('\n')[0])
    except NoSuchElementException:
        Price.append('-')
        
        
    try:
        replace=driver.find_element(By.XPATH,'//*[@id="iconfarmv2_feature_div"]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/ol/li[3]/div[1]/span/div[2]/span')
        Replacement.append(replace.text)
    except NoSuchElementException:
        Replacement.append('-')
        
        
        
    try:
        stock=driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-none a-spacing-top-micro }"]')
        Stock.append(stock.text)
    except NoSuchElementException:
        Stock.append('-')
        
        
    try:
        delivery=driver.find_element(By.XPATH,'//*[@id="deliveryBlockContainer"]/div[1]/div[1]/div[1]/div[1]/span/span[1]')
        Delivery.append(delivery.text)
    except NoSuchElementException:
        Delivery.append('-')
```

```python
print(len(Brand),len(Delivery),len(Stock),len(Price),len(Title),len(Replacement))
```

```python
df=pd.DataFrame({"Brand Name":Brand,"Name of the Product":Title,"Price":Price,"Return/Exchange":Replacement,"Expected Delivery":Delivery,"Availability":Stock,"Product URL":page_url})
df
```

```python
df.to_csv('question2.csv')
```
