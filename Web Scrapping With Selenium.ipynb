```python
import selenium 
import pandas as pd
from selenium import webdriver                   
from selenium.webdriver.common.by import By      
import warnings                                  
warnings.filterwarnings("ignore")
import time 
```

```python
driver = webdriver.Chrome()
```

```python
job_title=[]
company_name=[]
experience=[]
salary=[]
location=[]
```

```python
title=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title[0:10]:
    final=i.text
    job_title.append(final)
    
    
company=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company[0:10]:
    final1=i.text
    company_name.append(final1)
    
    
exp=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in exp[0:10]:
    final2=i.text
    experience.append(final2)
    
    
    
loc=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in loc[0:10]:
    final4=i.text
    location.append(final4)
```

```python
print(len(job_title),len(company_name),len(experience),len(location))
```
