# Question 1

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
Name=[]
Rank=[]
Artist=[]
Upload_date=[]
Views=[]
```

```python
try:
    rank=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody[1]/tr/td[1]')
    for i in rank[0:30]:
        Rank.append(i.text.split('.')[0])
        
except NoSuchElementException:
    Rank.append('-')
    

try:
    name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody[1]/tr/td[2]')
    for i in name[0:30]:
        Name.append(i.text.split('[')[0])
        
except NoSuchElementException:
    Name.append('-')
    

try:
    artist=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody[1]/tr/td[3]')
    for i in artist[0:30]:
        Artist.append(i.text.split(',')[0])
        
except NoSuchElementException:
    Artist.append('-')
    
    
try:
    upload=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody[1]/tr/td[5]')
    for i in upload[0:30]:
        Upload_date.append(i.text)
        
except NoSuchElementException:
    Upload_date.append('-')
    

try:
    views=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody[1]/tr/td[4]')
    for i in views[0:30]:
        Views.append(i.text)
        
except NoSuchElementException:
    Views.append('-')
```

```python
Views
```

```python
df=pd.DataFrame({"Rank":Rank,"Name":Name,"Artist":Artist,"Upload date":Upload_date,"Views in Billions":Views})
df
```

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
role13=driver.find_element(By.XPATH,'//a[@class="btn-primary btn"]')
role13.click()
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
Data_url=[]
```

```python
url=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in url:
    Data_url.append(i.get_attribute('href'))
```

```python
for i in Data_url:
    driver.get(i)
    
    try:
        dataset_name=driver.find_element(By.XPATH,'//main[@class="min-h-screen"]/div/div[1]/div[1]/div[1]/div[2]/div[1]/h1')
        Dataset_name.append(dataset_name.text)
        
    except NoSuchElementException:
        Dataset_name.append('-')
        
        
    try:
        data_type=driver.find_element(By.XPATH,'//main[@class="min-h-screen"]/div/div[1]/div[1]/div[2]/div[2]/div[1]/p[1]')
        Data_type.append(data_type.text)
        
    except NoSuchElementException:
        Data_type.append('-')
        
        
    try:
        task=driver.find_element(By.XPATH,'//main[@class="min-h-screen"]/div/div[1]/div[1]/div[2]/div[2]/div[3]/p[1]')
        Task.append(task.text)
        
    except NoSuchElementException:
        Task.append('-')
        
        
        
    try:
        attribute_type=driver.find_element(By.XPATH,'//main[@class="min-h-screen"]/div/div[1]/div[1]/div[2]/div[2]/div[4]/p[1]')
        Attribute_type.append(attribute_type.text)
        
    except NoSuchElementException:
        Attribute_type.append('-')
        
        
        
        
    try:
        no_of_instances=driver.find_element(By.XPATH,'//main[@class="min-h-screen"]/div/div[1]/div[1]/div[2]/div[2]/div[5]/p[1]')
        No_of_instances.append(no_of_instances.text)
        
    except NoSuchElementException:
        No_of_instances.append('-')
        
        
        
        
    try:
        no_of_attribute=driver.find_element(By.XPATH,'//main[@class="min-h-screen"]/div/div[1]/div[1]/div[2]/div[2]/div[6]/p[1]')
        No_of_attribute.append(no_of_attribute.text)
        
    except NoSuchElementException:
        No_of_attribute.append('-')
        
        
        
    try:
        year=driver.find_element(By.XPATH,'//main[@class="min-h-screen"]/div/div[1]/div[1]/div[1]/div[2]/h2')
        Year.append(year.text)
        
    except NoSuchElementException:
        Year.append('-')
```
